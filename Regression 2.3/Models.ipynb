## Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures,OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.compose import ColumnTransformer
from sklearn.metrics import r2_score, root_mean_squared_error
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

## Import Dataset

```python
df = pd.read_csv("used_cars.csv")
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['milage'] = df['milage'].str.replace('[^0-9]', '',regex=True).astype(int)
df= df.dropna()
```

```python
print(df.isnull().sum())
```

## Label Engineering

```python
df['engine_size'] = df['engine'].str.extract('(\d\.\d)').astype(float)
df['car_age'] = 2026 - df['model_year']
# handling missin values
df['fuel_type'].fillna('Unknown', inplace=True)
df['accident'].fillna('Unknown', inplace=True)
df['clean_title'].fillna('Unknown', inplace=True)
df = df.drop(columns=['model','ext_col','int_col','engine','model_year'])
rare_brands = df['brand'].value_counts()[df['brand'].value_counts() < 10].index

df['brand'] = df['brand'].replace(rare_brands, 'Other')
df = pd.get_dummies(df, drop_first=True)
x=df.drop(columns=["price"])
y =df["price"]
x.isna()
```

## Splitting Train/Test  data

```python
x_train,x_test,y_train ,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
def report(y_true,y_pred):
    return print(f"R2 score:{r2_score(y_true,y_pred)}\nRSME:{root_mean_squared_error(y_true,y_pred)}")
df.select_dtypes(include="object").columns
```

# Linear Regression model

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
lin_y_pred =lin_reg.predict(x_test)
report(y_test,lin_y_pred)
```

# SVR

```python
svm_reg = SVR(kernel='rbf')
svm_reg.fit(x_train,y_train)
svr_y_pred = svm_reg.predict(x_test)
```

# Decision Tree

```python
dt =DecisionTreeRegressor(random_state=0)
dt.fit(x_train,y_train)
dt_y_pred = dt.predict(x_test)
```

# Random Forest

```python
rf = RandomForestRegressor(n_estimators=150,random_state=0)
rf.fit(x_train,y_train)
rf_pred =rf.predict(x_test)
```

# XGBoost

```python
from xgboost import XGBRegressor
XGB = XGBRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='reg:squarederror',
    random_state=0)
XGB.fit(x_train,y_train)
xg_pred=XGB.predict(x_test)
```

```python
predictions= [("decision Tree",dt_y_pred),("Random Forest",rf_pred),("XGbosst",xg_pred)]
```

```python
for name, preds in predictions:
    print(f"{name} Report")
    report(y_test, preds)
```
